```python
import matplotlib.pyplot as plt
import numpy as np
import caiman as cm
import bokeh.plotting as bpl
import holoviews as hv
from pathlib import Path

bpl.output_notebook()
hv.notebook_extension("bokeh")
```

```python
control_path = Path(
    "/workspaces/05_Caiman_Analysis/data/2024-07-17_RCX_div36_ctrl_NBplus_Calbryte-520"
)

lsd_path = Path(
    "/workspaces/05_Caiman_Analysis/data/2024-07-18_RCX_div37_LSD-1uM_NBplus_Calbryte-520"
)
```

```python
# give me every video with *__FOV001_Analog* in the name
movie_files = list(Path(control_path).rglob("*__FOV002*.tif"))
movie_files = sorted(movie_files)

movie_files
```

```python
# Load the movies and calculate the correlation images
correlation_images = []
for movie_path in movie_files:
    movie = cm.load(movie_path)
    correlation_image = cm.local_correlations(movie, swap_dim=False)
    # Get rid of NaNs, if they exist
    correlation_image[np.isnan(correlation_image)] = 0
    correlation_images.append(correlation_image)
```

```python
# Plot the correlation images side by side
fig, axes = plt.subplots(2, int(len(correlation_images) / 2), figsize=(18, 12))
axes = axes.flatten()

for ax in axes[len(correlation_images):]:
    ax.axis("off")

for ax, correlation_image, movie_path in zip(axes, correlation_images, movie_files):
    ax.imshow(
        correlation_image,
        cmap="viridis",
        vmin=np.percentile(np.ravel(correlation_image), 50),
        vmax=np.percentile(np.ravel(correlation_image), 99.5),
    )
    ax.set_title(f'{movie_path.name.split("__")[1]}', fontsize=12)

fig.suptitle(movie_path.name.split("__")[0], fontsize=12, y=1.02)
plt.tight_layout()
plt.show()
```
